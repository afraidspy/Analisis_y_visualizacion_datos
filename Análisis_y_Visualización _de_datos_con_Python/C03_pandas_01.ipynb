Agregar columnas a un daframe

Agregar una columna con una etiqueta: Si la pelicula tiene calificacion >= 4 entonces colocar Recomendable
                                      Si la pelicula tiene calificacion < 4 entonce colocar No recomendable


```python
tabla_peliculas["etiqueta_calificacion"] = tabla_peliculas["calificacion"] >= 4
tabla_peliculas["etiqueta_calificacion"] 
```

```python
# 1. Crear una función
def crear_etiqueta(calificacion):
    """if calificacion >= 4:
        return "Recomendable"
    else:
        return "No recomendable"
        """
    return  "Recomendable" if calificacion >= 4 else "No recomendable"

print("Etiqueta: " , crear_etiqueta(5))
    
# 2.Aplicar esa función a una columna en específico
tabla_peliculas["etiqueta_recomendacion"] =  tabla_peliculas["calificacion"].apply(crear_etiqueta)
```

```python
tabla_peliculas
```

Crear una columna con una etiqueta: si el anio > 2000 entonces poner reciente, en otro caso colocar antigua


```python
# 1. Crear función
def crear_etiqueta_anio(anio):
    return "Reciente" if anio > 2000 else "Antigua"
# Se valida que funcione correctamente crear_etiqueta_anio
crear_etiqueta_anio(1800)   
# 2. Aplicar la funcion a la columna anio
tabla_peliculas['etiqueta_antiguedad'] = tabla_peliculas['anio'].apply(crear_etiqueta_anio)
```

## Fechas

```python
tabla_ventas = pd.DataFrame(
    {
    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Audífonos"],
    "fecha_venta": ["2025-01-10", "2025-02-15", "2025-03-01", "2025-03-20", "2025-04-05"],
    "monto": [18000, 450, 950, 4200, 1200]
})

print(tabla_ventas)
```

```python
print(tabla_ventas.info())
```

```python
print(tabla_ventas["fecha_venta"].dtypes)
```

```python
tabla_ventas["fecha_venta"] = pd.to_datetime(tabla_ventas["fecha_venta"])
```

```python
print(tabla_ventas["fecha_venta"].dtypes)
```

```python
tabla_ventas["dia"] = tabla_ventas["fecha_venta"].dt.day
tabla_ventas
```

```python
tabla_ventas["mes"] = tabla_ventas["fecha_venta"].dt.month
tabla_ventas
```

```python
tabla_ventas["anio"] = tabla_ventas["fecha_venta"].dt.year
tabla_ventas
```

```python
tabla_ventas["trimestre"] = tabla_ventas["fecha_venta"].dt.quarter
tabla_ventas
```

```python
tabla_ventas["dia_semana"] = tabla_ventas["fecha_venta"].dt.dayofweek
tabla_ventas
```

```python
tabla_ventas["nombre_dia"] = tabla_ventas["fecha_venta"].dt.day_name()
print(type(tabla_ventas["dia_semana"]))
tabla_ventas.iloc[0]["dia_semana"]
```

```python
tabla_ventas
```

```python
nombres_dias = {
    0: "lunes",
    1: "martes",
    2: "miércoles",
    3: "jueves",
    4: "viernes",
    5: "sábado",
    6: "domingo"
}

nombres_meses = {
    1: "enero",
    2: "febrero",
    3: "marzo",
    4: "abril",
    5: "mayo",
    6: "junio",
    7: "julio",
    8: "agosto",
    9: "septiembre",
    10: "octubre",
    11: "noviembre",
    12: "diciembre"
}


## Mapeo de valores para hacer la sustición 
tabla_ventas["nombre_dia"] = tabla_ventas["dia_semana"].map(nombres_dias)
tabla_ventas["nombre_mes"] = tabla_ventas["mes"].map(nombres_meses)
tabla_ventas
```

El monto total de todas las ventas del mes marzo

```python
filtro =  tabla_ventas["mes"] == 3
filtro 
tabla_ventas[filtro]["monto"].sum()
```

Obtener el mes con mayor ganancia

```python
tabla_ventas.groupby("nombre_mes")["monto"].sum()
tabla_ventas.groupby("nombre_mes")["monto"].sum().reset_index().sort_values("nombre_mes").max()
```

```python
tabla = pd.pivot_table(tabla_ventas, index='nombre_mes', values='monto' , aggfunc='sum')
tabla.sort_values(by='monto', ascending = False).head(1)
```

```python
tabla_ventas.groupby("nombre_mes")["monto"].sum().idxmax()
```
