```python
#df["categorias_genero"] = df["Genre"]



df["categorias_freq"] = df["Genre"].map(df.groupby("Genre")["Genre"].count())


# Conndición 1: 
#df[["Genre","categorias_genero"]]
#condicion = (df["Genre"].value_counts() / len(df) ) *  100
df["categoria_nueva"] = np.where( df["categorias_freq"]  <= 2, "Otros(2%)" , df["categorias_freq"])

df["categoria_etiquetas"] =  np.where( df["categoria_nueva"]== "Otros(2%)" , df["categoria_nueva"], df["Genre"] ) 
df
```

```python
plt.figure(figsize=(8, 8))
plt.title("Distribución de canciones por género agrupando categoría <2%")

plt.pie(df["categoria_etiquetas"].value_counts(),
        labels = df["categoria_etiquetas"].unique(),
        autopct="%1.0f%%",
        startangle=90)

plt.title("Distribución de cancione por género")

plt.show()
```

## Correlación

```python
columnas_numericas = [
    "Beats.Per.Minute",
    "Energy",
    "Danceability",
    "Loudness..dB..",
    "Liveness",
    "Valence.",
    "Length.",
    "Acousticness..",
    "Speechiness.",
    "Popularity"
]
matriz_correlacion = df[columnas_numericas].corr()

matriz_correlacion
```

```python
matriz_correlacion.shape
```

- Valores cercanos a 1 indican correlación positiva fuerte.
- Valores cercanos a -1 indican correlación negativa fuerte.
- Valores cercanos a 0 indican poca relación lineal.

```python
# Seleccionar los 5 géneros musicales más frecuentes en el dataset
generos_principales = df["Genre"].value_counts().head(5).index

# Filtrar el DataFrame para conservar únicamente las canciones de esos géneros
df_filtrado = df[df["Genre"].isin(generos_principales)]

# Preparar los datos que se usarán en el boxplot
# Se crea una lista donde cada elemento contiene la popularidad de un género
datos_boxplot = [
    df_filtrado[df_filtrado["Genre"] == genero]["Popularity"]
    for genero in generos_principales
]

# Crear una figura de 10 pulgadas de ancho por 6 pulgadas de alto
plt.figure(figsize=(10, 6))

# Crear un boxplot para comparar la distribución de popularidad por género
plt.boxplot(datos_boxplot, labels=generos_principales)

# Agregar título a la gráfica
plt.title("Distribución de popularidad por género")

# Agregar etiqueta al eje X
plt.xlabel("Género")

# Agregar etiqueta al eje Y
plt.ylabel("Popularidad")

# Rotar las etiquetas del eje X para que se lean mejor
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

## Personalización de gráficas

```python
# Popularidad pronedio de cada género musica

popularidad_promedio = (
    df.groupby("Genre")["Popularity"]
    .mean()
    .sort_values(ascending = False)
    .head()

)

plt.figure(figsize=(12, 6))

plt.bar(
    popularidad_promedio.index,         # Nombres de los géneros musicales
    popularidad_promedio.values,
    color =  "#CD1C95"         # Valores de popularidad promedio

)


# Agregar título a la gráfica
plt.title("Popularidad promedio por género musical", fontsize=14)

# Agregar etiqueta al eje X
plt.xlabel("Género")

# Agregar etiqueta al eje Y
plt.ylabel("Popularidad promedio")

# Rotar las etiquetas del eje X para que se lean mejor
plt.xticks(rotation=45, ha="right")

# Agregar una cuadrícula horizontal para facilitar la lectura de los valores
plt.grid(alpha=0.5)


for indice, valor in enumerate(popularidad_promedio.values):
    plt.text(
        indice,                         # Posición horizontal de la etiqueta
        valor + 1,                    # Posición vertical ligeramente arriba de la barra
        round(valor, 1),                # Texto que se muestra: valor redondeado a 1 decimal
        ha="center"                     # Alinear el texto al centro de la barra
    )


#plt.grid(axis="x",alpha=0.5)


plt.show()


```
