## Imputación de la columna precio
Se usa el promedio por que suaviza los precios,
Desicion: Remplazo por promedio por colonia y tipo de propiedad para tener un precio más acorde a la ubicación, ya que como se analizo este depende de la colonia y de tipo de propiedad.


```python
df_copia["precio_agregado"] = df_copia["precio"]
df_precios = df_copia.groupby("colonia")["precio"].mean()
df_precios


```

```python
## Diccionario con promedios por cada colonoa --> Python
## Groupby

df_copia["precio_agregado"] = np.where(df_copia["precio_agregado"].isna(), df_copia["colonia"].map(df_precios), df_copia["precio_agregado"])
```
