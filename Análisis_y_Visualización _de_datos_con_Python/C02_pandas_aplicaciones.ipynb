## Series

```python
serie_ventas= pd.Series([120, 150, 180, 170], 
                         index=["Monday", "Tuesday", "Wednesday", "Thursday"])
```

```python
serie_ventas
```

```python
print("Valores: ", serie_ventas.values)
print("Tipo de datos: ", type(serie_ventas.values))
```

```python
print("Indice: ", list(serie_ventas.index))
```

```python
print("Tipo dato: ", serie_ventas.dtype)
```

```python
print("Cantidad de elementos de una serie: ", serie_ventas.size)
```

```python

print("¿Cuál es el valor de Monday? ", serie_ventas["Monday"])
```

```python
print("Subconjunto: \n", serie_ventas[0:2]) # Subconjunto con los elementos de las posiciones 0 y 1
```

```python
empleados_empresa = pd.DataFrame({
    "empleado": ["E001", "E002", "E003", "E004", "E005"],
    "departamento": ["Ventas", "Finanzas", "Ventas", "Operaciones", "Finanzas"],
    "salario": [2200, 3100, 2400, 2800, 3300],
    "antiguedad_anios": [2, 5, 3, 4, 6]
})

empleados_empresa
```

```python
empleados_empresa.shape
```

```python
empleados_empresa.head() #Regresa los primeros 5 elementos del dataframe
```

```python
empleados_empresa.head(2) 
```

```python
renombre_empleados_df =  empleados_empresa.rename(
    columns = {
        "empleado": "Empleado",
        "departamento": "Departamento",
        "salario" : "Salario",
        "antiguedad_anios": "Antigüedad"
    }
)

#cve_pais 
```
