## Aplanado de un arreglo

````python
mediciones_horas = np.array([
    [12, 15, 11, 14], #Mexico
    [18, 17, 16, 19], #Colobia
    [10, 12, 13, 11]  #Argentina
])

aplanado_horas = mediciones_horas.flatten()
print("Uso de flatten \n", aplanado_horas)
print("Dimensiones \n", aplanado_horas.shape)
print("Tamaño \n", aplanado_horas.size)
````

````python
aplando_horas = mediciones_horas.ravel()
print("Uso de flatten \n", aplanado_horas)
print("Dimensiones \n", aplanado_horas.shape)
print("Tamaño \n", aplanado_horas.size)
````

## Combinación de arreglos

````python
valores_a = np.array([21.5, 22.0, 21.8, 22.4, 10, 2])
valores_b = np.array([22.1, 22.3, 22.0, 21.9])
````

````python

temperaturas = np.concatenate([valores_a, valores_b])
print("Concatenación de temperaturas: ",temperaturas)
print("Size: ", temperaturas.size)
````

````python
region_norte = np.array([
    [15, 18, 17],
    [20, 22, 21]
])

region_sur = np.array([
    [25, 24, 26],
    [19, 18, 20]
])
````

````python
apilado_vertical = np.vstack([region_norte, region_sur]) # Unión de dos matrices de forma vertical
print("Apilado vertical: \n", apilado_vertical)
print("Sahpe: ", apilado_vertical.shape)
````

````python
turno_manana = np.array([
    [42, 38],
    [35, 40],
    [50, 47]
])

turno_tarde = np.array([
    [39, 41],
    [36, 37],
    [48, 46]
])
````

````python
apilado_horizontal = np.hstack([turno_manana, turno_tarde])
print("Matriz apilada horizontalmente: \n",apilado_horizontal)
print("Shape: ", apilado_horizontal.shape)
````

Nota: 
 - vstack(): Va aumentar el número de filas
 - hstack(): Va aumentar el número de columnas

## Dividir un arreglo (Split)

````python
edad = np.array([11, 13, 15, 14, 16, 18, 17, 19, 21, 20, 22, 24])

print("Edad :\n", edad )
print("Tamanño: ", edad.size)
````

````python
arreglo_dividido = np.split(edad, 4) #[array([11, 13, 15]), array([14, 16, 18]), array([17, 19, 21]), array([20, 22, 24])]
print(arreglo_dividido)
````

````python
print("Tipo del arreglo_dividido: ", type(arreglo_dividido))
````

````python
primer_arreglo = arreglo_dividido[0] #Corresponde al primer arreglo de la lista que es: [11 13 15]
print("El primer arreglo es: ", primer_arreglo)
print("Tipo de dato del primer arreglo ", type(primer_arreglo))
````

````python
segundo_arreglo = arreglo_dividido[1] #Corresponde a la segunda posición del arreglo
print("El segundo arreglo es: ", segundo_arreglo ) 
print("Tipo de dato del segundo arreglo ", type(segundo_arreglo ))
````

````python
tercer_arreglo = arreglo_dividido[2] #Corresponde a la segunda posición del arreglo
print("El tercer arreglo es: ", tercer_arreglo ) 
print("Tipo de dato del tercer arreglo ", type(tercer_arreglo))
````

### Ejercicio 01
Usa broadcasting para sumar el vector `[5, 10, 15]` a cada fila de la matriz siguiente:

```python
[[20, 25, 30],
 [35, 40, 45]]
```

````python
matriz_base = np.array([
    [20, 25, 30],
    [35, 40, 45]
])

#incremento = np.array([5, 10, 15,10])
incremento = np.array([5, 10, 15])

tupla = matriz_base.shape
print("Tipo de datos: ", type(tupla))
numero_columnas_matriz_base = tupla[1] # 1 corresponde al número de columnas
print(numero_columnas_matriz_base)

if (tupla[1] == incremento.size ): # Número de columnas == cantidad de elementos del arreglo
    suma_vector = matriz_base + incremento
    print(suma_vector)
else:
    print("No se puede efectuar la operación")
````

````python
def suma_broadcasting(matriz ,  vector):
    tupla = matriz.shape

    if (tupla[1] == vector.size):
        suma_vector = matriz + vector
        return suma_vector
    else: 
        print("No se peude efectuar la operación")
        return None
    
````

````python
matriz_base = np.array([
    [20, 15, 3],
    [75, 30, 5]
])
incremento = np.array([5, 10, 15])

# Mandar a llamar a la función suma_broadcasting
suma_broadcasting(matriz_base, incremento)
````
