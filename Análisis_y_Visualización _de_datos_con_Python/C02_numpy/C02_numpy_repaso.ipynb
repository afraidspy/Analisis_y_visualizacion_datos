## 1. ¿Qué es un arreglo de NumPy?

Un arreglo de NumPy es una estructura diseñada para almacenar datos numéricos de forma eficiente.  
A diferencia de una lista de Python, un arreglo suele trabajar mejor cuando todos sus elementos tienen el mismo tipo y cuando queremos hacer cálculos sobre muchos valores a la vez.



```python
valores = np.array([18.5, 22.1, 19.8, 25.4, 21.0, 17.6])

print("Arreglo:", valores)
print("Tipo del objeto:", type(valores))
print("Tipo de los datos:", valores.dtype)
```

```python
valores = np.array(["Pedro", True, 20, 10.0])
print(valores.dtype)
print(valores)
```

## 2. Propiedades básicas

Para conocer conocer la estructura del arreglo


- `shape`: forma del arreglo
- `ndim`: número de dimensiones
- `size`: número total de elementos
- `dtype`: tipo de dato almacenado


```python
#Cada valor de la matrix representa un nivel de contaminación del aire
calidad_aire = np.array([
    [42, 38, 35, 40],
    [55, 60, 58, 62],
    [30, 28, 27, 31]
])

print("Matriz de calidad del aire:")
print(calidad_aire)
print("shape:", calidad_aire.shape) #Se accede a una propiedad del arreglo
print("ndim:", calidad_aire.ndim)
print("size:", calidad_aire.size)
print("dtype:", calidad_aire.dtype)
```

Esta matriz tiene:

- 3 filas: tres ciudades
- 4 columnas: cuatro valores que representan que tan contaminada esta esa ciudad.

## 4. Acceso a elementos y slicing

El acceso por índice nos permite extraer un valor específico, una fila, una columna o un bloque del arreglo.

En este ejemplo, cada fila representa una ruta de reparto y cada columna representa un tramo del recorrido medido en kilómetros.

```python
rutas_reparto = np.array([
    [4.2, 5.1, 3.8, 6.0], #fila representa una ruta de reparto
    [7.5, 2.4, 4.1, 5.6], #columna: tramo del recorrido
    [3.3, 3.9, 4.8, 2.7]
])

print("Matriz completa:")
print(rutas_reparto)
#print("\nPrimer valor de la primera ruta:", rutas_reparto[0, 0])
#print("Último tramo de la segunda ruta:", rutas_reparto[1, -1]) # Fila 1 , Columna: -1 -a la última
#print("Primera ruta completa:", rutas_reparto[0])
#print("Segunda ruta completa:", rutas_reparto[1])
#print("Tercer ruta completa:", rutas_reparto[2])
#print("Todos los primeros tramos:", rutas_reparto[:, 0]) #[ filas, columnas]
#print("Todos los primeros tramos:", rutas_reparto[0:3, 0:2]) #[ posisicones de las filas, posiciones de las columnas]
#?# rutas_reparto[0:3, 0:2]
print("Submatriz con las dos primeras rutas y los dos últimos tramos:")
# Tomar las primeras 2 filas
# y desde la tercera columna hasta el final
print(rutas_reparto[0:2, 2:4])  # Notación más explícita
print(rutas_reparto[0:2, -2:])  #AlternativaTener cuidado con los -
print(rutas_reparto[:2, 2:])   # [desdel inicio hasta la posicion 1, desde la posición 2 hasta el resto]
```


- `arreglo[fila, columna]` accede a un elemento exacto
- `arreglo[0]` devuelve la primera fila
- `arreglo[:, 0]` devuelve la primera columna
- `arreglo[inicio:fin]` permite tomar un rango

## 6. Operaciones estadísticas

NumPy incluye funciones datos rápidamente.

En el siguiente ejemplo analizamos el número de visitantes en un museo, donde cada fila representa una sala y cada columna una hora del día.

```python
visitantes_museo = np.array([
    [1,  1, 49, 60],  #Fila: sala
    [30, 28, 35, 42],  #Columna: hora del dia
    [55, 58, 61, 66]
])
#print("Total general:", np.sum(visitantes_museo))
#print("Promedio general:", np.mean(visitantes_museo))
#print("Valor mínimo:", np.min(visitantes_museo))
#print("Valor máximo:", np.max(visitantes_museo))
#print("Desviación estándar:", np.std(visitantes_museo))
print("\nTotal por sala (filas):", np.sum(visitantes_museo, axis=1)) #Suma de filas  [1 + 1 + 49 + 60, 30+28+35+42, 55+58+61+66]
print("Total por hora (columnas):", np.sum(visitantes_museo, axis=0)) # Suma por columnas [1+30+55, 1+28+58, 49+35+61, 60+42+66]
print("Promedio por sala:", np.mean(visitantes_museo, axis=1)) #Promedio por filas
print("Promedio por hora:", np.mean(visitantes_museo, axis=0)) # Promedio por columnas
```

La desviación estándar mide la dispersión de los datos.
Entre mayor sea, más separados están de la media; entre menor sea, más cerca están del promedio.

`axis`
- `axis=0`: recorre hacia abajo y resume por columnas
- `axis=1`: recorre hacia la derecha y resume por filas



## 7. Filtros booleanos

Los filtros booleanos sirven para seleccionar solo los datos que cumplen una condición.

Trabajaremos con tiempos de respuesta de un servicio en línea medidos en milisegundos.

```python
tiempos_respuesta = np.array([180, 220, 145, 310, 275, 198, 405, 167, 250])

filtro_respuestas_lentas = tiempos_respuesta > 250
filtro_ventana_normal = (tiempos_respuesta >= 150) & (tiempos_respuesta <= 250)

print("Filtro de respuestas lentas:", filtro_respuestas_lentas)
print("Respuestas lentas:", tiempos_respuesta[filtro_respuestas_lentas])

print("Respuestas dentro del rango normal:", tiempos_respuesta[filtro_ventana_normal])
print("Clasificación con np.where:")
print(np.where(tiempos_respuesta > 300, "critico", "aceptable")) # (filtro o condicion, condicion = True coloco "Critico", condicion = False coloco Aceptable)
```

### Nota
Primero generamos una condición que produce `True` o `False`.  
Después usamos esa condición para quedarnos solo con los valores que nos interesan.

## 8. Manejo de valores faltantes con `NaN`

En datos reales es muy común encontrar registros faltantes.  
NumPy representa estos valores como `NaN` (`Not a Number`).

En este ejemplo, una red de sensores registra niveles de humedad, pero algunos sensores fallaron temporalmente.

```python
humedad = np.array([58.0, 61.5, np.nan, 63.2, 59.8, np.nan, 60.1])

print("Datos originales:", humedad)
print("¿Dónde hay NaN?:", np.isnan(humedad))
print("Suma ignorando NaN:", np.nansum(humedad))
print("Promedio ignorando NaN:", np.nanmean(humedad))
print("Mínimo ignorando NaN:", np.nanmin(humedad))
print("Máximo ignorando NaN:", np.nanmax(humedad))
print("Reemplazo de NaN por 0:", np.nan_to_num(humedad, nan=0))
```

## 9. Reemplazo condicional con `np.where`
Remplazar los nan por el promedio

```python
promedio_humedad = np.nanmean(humedad)
humedad_completada = np.where(np.isnan(humedad), promedio_humedad, humedad)

print("Promedio usado para reemplazar:", promedio_humedad)
print("Serie completada:", humedad_completada)
```
