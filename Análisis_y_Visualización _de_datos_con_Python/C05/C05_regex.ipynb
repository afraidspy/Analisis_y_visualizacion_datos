## Registros con correos electrónicos no válidos

```python
filtro = ~clientes["correo"].str.contains(r"[\w._]+@[\w._]+", na= False)  # Negación en filtro en pandas usar ~
filtro
```

```python
clientes[filtro]
```

```python
# \w = [A-Za-z0-9]
filtro = clientes["correo"].str.contains(r"[\w._]+@[\w._]+", na= False)  # Negación en filtro en pandas usar ~
# * 0 o más elementos
# + 1 o más elementos
clientes[~filtro]
```

```python
filtro = clientes["correo"].str.contains(r"[\w._]{3}@[\w._]+", na= False)  # Negación en filtro en pandas usar ~
# * 0 o más elementos
# + 1 o más elementos
clientes[~filtro]
```

```python
import re ## Biblioteca pora manejar patrones en cadenas

pattern = "cliente"
text = "CliEnte is powerful. python is versatile."
matches = re.findall(pattern, text, flags=re.IGNORECASE)
print(matches) 
```

```python
def encontrar_patron(comentario):
    return re.findall("cliente", comentario, flags=re.IGNORECASE )
```

```python
clientes["comentario"].apply(encontrar_patron)
```
