## Imputación de la columna `precio`

La columna `precio` tiene valores faltantes. Como el precio de un alojamiento depende mucho de la ubicación(dada por la ubicaciòn) y del tipo de habitación, Por lo tanto no serìa adecuado que se imputen  los  valores faltantes con el promedio general, ya que meterìa sesgo en nuestros datos.

Despuès de hacer un anàlisis en las variables, podemos 

1. Primero, mediana por `colonia` y `tipo_habitacion`
2. Si todavía aùn hay faltantes, mediana por `colonia`
3. Si todavía hay faltantes, mediana por `tipo_habitacion`
4. Finalmente se puede usar mediana general del dataset

Se usa la **mediana** porque el precio tiene valores atípicos muy altos; la media puede deformarse por esos casos extremos.


````python
# 1. Contar valores faltantes antes de la imputación
faltantes_antes = df_copia["precio"].isna().sum()
faltantes_antes
````

````python
porcentaje_faltantes_antes = df_copia["precio"].isna().mean() * 100
porcentaje_faltantes_antes
````

````python
df_copia["precio_imputado"] = df_copia["precio"]
````

`transform()` calcula un resumen por grupo, pero devuelve un resultado del mismo tamaño que el DataFrame original.

Por ejemplo, si una fila pertenece a la colonia Roma y al tipo `Alojamiento entero`, `transform("median")` coloca en esa fila la mediana del precio de ese grupo. Por eso se puede usar directamente dentro de `fillna()`.

````python
df_copia[[
    "colonia",
    "tipo_habitacion",
    "precio",
    "precio_imputado"
]].head(10)
````

````python
# 1. Calcular la mediana del precio para cada colonia y tipo de habitación
mediana_por_grupo = df_copia.groupby(
    ["colonia", "tipo_habitacion"]
)["precio"].transform("median")

mediana_por_grupo
````

````python
#  Rellenar los precios faltantes usando esa mediana
df_copia["precio_imputado"] = df_copia["precio_imputado"].fillna(
    mediana_por_grupo
)
df_copia["precio_imputado"] 
````

````python
faltantes_despues_nivel_1 = df_copia["precio_imputado"].isna().sum()

faltantes_despues_nivel_1
````

## Imputar los faltantes restantes usando la mediana por tipo de habitación

````python

df_copia["precio_imputado"] = df_copia["precio_imputado"].fillna(
    df_copia.groupby("tipo_habitacion")["precio"].transform("median")
)

# Contar cuántos valores siguen faltando después del nivel 3
faltantes_despues_nivel_3 = df_copia["precio_imputado"].isna().sum()

faltantes_despues_nivel_3
````

## Funciòn `agg()` : Signfifica agregar o resumir datos

## Multiíndices en Pandas



Un **multiíndice** es un índice con varios niveles.

En Pandas aparece comúnmente cuando agrupamos los datos por más de una columna usando `groupby()`.

## Idea principal

Cuando agrupamos por una sola columna, el índice tiene un solo nivel.

## Uso de Pivot

## ¿Qué es una tabla dinámica o `pivot_table`?

Una tabla dinámica permite resumir y reorganizar un DataFrame para analizar la información de forma cruzada.

En lugar de ver los datos registro por registro, una tabla dinámica nos permite responder preguntas como:

**¿Cuál es el precio promedio de los alojamientos por colonia y tipo de habitación?**

En el dataset de Airbnb, cada fila representa un alojamiento. Algunas columnas importantes son:

- `colonia`: zona donde se encuentra el alojamiento.
- `tipo_habitacion`: tipo de alojamiento, por ejemplo habitación privada o departamento completo.
- `precio`: precio del alojamiento.

Con `pivot_table`, podemos convertir esta información en una tabla resumen donde:

- Las filas representan las colonias.
- Las columnas representan los tipos de habitación.
- Los valores representan el precio promedio.

La estructura general es:

```python
pd.pivot_table(
    data=df_copia,
    index="colonia",
    columns="tipo_habitacion",
    values="precio",
    aggfunc="mean"
)

````python
pd.pivot_table(
    data=df_copia,
    index="colonia",
    columns="tipo_habitacion",
    values="precio_imputado",
    aggfunc="mean"
)
````

Esta tabla dinámica muestra el precio promedio de los alojamientos de Airbnb según dos variables:

- Filas: `colonia`
- Columnas: `tipo_habitacion`
- Valores: promedio de `precio`

Cada celda representa el precio promedio para una combinación específica.

Por ejemplo:

En `Benito Juárez`, el precio promedio de `Entire home/apt` es aproximadamente 1444.69, mientras que el de `Private room` es 657.92.

Los valores `NaN` significan que no existen registros para esa combinación. Por ejemplo, si aparece `NaN` en `Hotel room` para una colonia, significa que en esa colonia no hay alojamientos registrados de ese tipo.

Esta tabla permite comparar rápidamente qué colonias son más caras y cómo cambia el precio según el tipo de habitación.
