```python
import pandas as pd
```

```python
estudiantes = pd.DataFrame({
    "id_estudiante": [1, 2, 3, 4, 5], # Llave primaria : valor único (PK: Primary Key)
    "nombre": ["Ana", "Luis", "Marta", "Carlos", "Sofía"],
    "grupo": ["A", "A", "B", "B", "C"]
})

estudiantes
```

```python
calificaciones = pd.DataFrame({
    "id_estudiante": [1, 2, 3, 6],  # Llave foránea (Foreign Key:  FK)
    "id_curso": [101, 101, 102, 103],  # llave primaria (PK)
    "calificacion": [9.5, 8.0, 10.0, 7.5]
})

calificaciones
```

```python
cursos = pd.DataFrame({
    "id_curso": [101, 102, 103, 104], # Primary key
    "curso": ["Python", "Pandas", "Bases de datos", "Estadística"],
    "profesor": ["Dr. Gómez", "Mtra. Ruiz", "Dr. Pérez", "Mtra. Torres"]
})

cursos
```

# Uso de merge

-- how : left, right, inner, outer


```python
## Dataframe izquierdo: esrtudiantes
## Dataframe derecho calificaciones

# Left: Se mantienen todas las filas del dataframe izquierdo
estudiantes.merge(
    calificaciones,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "left"
)
```

```python
calificaciones.merge(
    estudiantes,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "left"
)
```

```python
# Right: Mantienen todas las filas del dataframe derecho
estudiantes.merge(
    calificaciones,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "right"
)
```

```python
calificaciones.merge(
    estudiantes,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "inner"
)
```

```python
estudiantes
```

```python
calificaciones
```

```python
# Inner: Conserva las filas o registros que tengan coincidencia en ambos dataframes o tabla

estudiantes.merge(
    calificaciones,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "inner"
)
```

```python
calificaciones.merge(
    estudiantes,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "inner"
)
```

```python
# Outer: Conserva las filas de ambas tablas (dataframes): Tanto los de la izquierda como de la derecha
estudiantes.merge(
    calificaciones,
    on = "id_estudiante", # Cuál es la columna que existe en ambos dataframe,
    how = "outer"
)
```

```python
calificaciones.merge(
    estudiantes,
    on = "id_estudiante",
    how = "outer"
)
```

## JOIN

- Permite unir tablas o dataframes, pero utilizando el índice de este

```python
# Indicamos explícitamente que una columna debe de ser el índice del dataframe
estudiantes_inidice = estudiantes.set_index("id_estudiante") # Se ocuparía la llave primaria
calificaciones_indice = calificaciones.set_index("id_estudiante")

estudiantes_inidice
```

```python
calificaciones_indice
```

```python
union_join = estudiantes_inidice.join(
    calificaciones_indice,
    how = "left"
)

union_join 
```

```python
union_join = estudiantes_inidice.join(
    calificaciones_indice,
    how = "right"
)

union_join 
```

```python
union_join = estudiantes_inidice.join(
    calificaciones_indice,
    how = "inner"
)

union_join 
```

```python
union_join = estudiantes_inidice.join(
    calificaciones_indice,
    how = "outer"
)

union_join 
```

## Concat

```python
grupo_a = pd.DataFrame({
    "id_estudiante": [1, 2],
    "nombre": ["Ana", "Luis"],
    "grupo": ["A", "A"]
})

grupo_b = pd.DataFrame({
    "id_estudiante": [3, 4],
    "nombre": ["Marta", "Carlos"],
    "grupo": ["B", "B"]
})
```

```python
grupo_a
```

```python
grupo_b
```

```python
estudiantes_todos =  pd.concat(
    [grupo_a, grupo_b],
    axis = 0  # Unión a partir de columnas, colocar un dataframe abajo de otro

)

estudiantes_todos
```

```python
estudiantes_todos =  pd.concat(
    [grupo_a, grupo_b],
    axis = 1  # Unión a partir de filas, une el segundo dataframe  a la derecha del primero

)

estudiantes_todos
```
