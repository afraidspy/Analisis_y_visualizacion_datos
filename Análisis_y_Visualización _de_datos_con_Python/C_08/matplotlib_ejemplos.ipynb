```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo"]
ventas = [120, 150, 90, 180, 210]

plt.plot(meses, ventas)  # crear la gráfica
plt.show() # asegurarse de que se muetstre sagráfoca
```

```python
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo"]
ventas = [120, 150, 90, 180, 210]

plt.figure(figsize = (10, 8))
plt.plot(meses,# crear la gráfic, lineal
        ventas,
        marker = "o",
        linestyle = "--",
        color = "orange",
        linewidth = 2
        )  
plt.title("Ventas mensuales de 5 meses", fontsize = 16)

plt.xlabel("Meses")
plt.ylabel("Ventas")

plt.grid(True)
plt.show() # asegurarse de que se muetstre sagráfoca
```

```python
df_original =  pd.read_csv("top50.csv", encoding="latin1")

df_original
```

```python
df = df_original.copy()
```

```python
df.shape
```

```python
df.info()
```

```python
generos = df["Genre"].value_counts().sort_values(ascending = False)
generos
```

```python
plt.figure(figsize = (20,30))
plt.title("Distribución de géneros en el 2019")
plt.bar(generos.index, generos.values )
plt.xticks(ticks=generos.index,  rotation=45, ha='right')
plt.show()
```

## Validar la popularidad

```python
df.sort_values(by = "Popularity", ascending  = False)

plt.bar(df["Track.Name"], df["Popularity"])

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["Popularity"], bins = 10, color = "green")
plt.title("Diagrama de distrución de la popularidad de canciones en el 2019")
plt.xlabel("popularidad")
plt.ylabel("frecuencia")

plt.show()
```
